```python
import plotly.figure_factory as ff
import numpy as np

Co_s, Ni_s = np.mgrid[0:1:14j, 0:1:14j]
Co_s, Ni_s = Co_s.ravel(), Ni_s.ravel()
mask = Co_s + Ni_s <= 1
Co_s, Ni_s = Co_s[mask], Ni_s[mask]
Fe_s = 1 - Co_s - Ni_s
Fe_s[Fe_s < 0] = 0 

overpotential_min = 100
overpotential_max = 2000
for time in [2, 24]:
    for temp in [300, 800]:
        print("t: {}, T: {}".format(time, temp))


        values = 100 - (abs((Co_s-0.35)*10)**2 + abs((Ni_s-0.5)*10)**2 + ((time-8)*2)**2 + ((temp - 500)/10)**2) 
        values = values + (abs((Co_s-0.35)*10)**3 + abs((Ni_s-0.5)*10)**3 + ((time-8))**4 + ((temp - 500)/10)**4)/4
        values = values - Fe_s * (time-8)**2 - Co_s * (time - 4)**2 - Co_s * Ni_s * (temp - 500)/10
        values = values/2000

        
        
        points = np.c_[Co_s, Ni_s, [time] * len(Co_s), [temp] * len(Co_s)]
 
        
        fig = ff.create_ternary_contour(np.array([Co_s, Fe_s, Ni_s]), values,
                                pole_labels=['Co', 'Fe', 'Ni'],
                                interp_mode='cartesian',
                                ncontours= 20,
                                colorscale= 'Jet',
                                showscale=True,
                                showmarkers=True,
                                title='Overpotential, Time = ' + str(time) + ', Temp = ' + str(temp))
        
       
        fig.show()
        # break
```
